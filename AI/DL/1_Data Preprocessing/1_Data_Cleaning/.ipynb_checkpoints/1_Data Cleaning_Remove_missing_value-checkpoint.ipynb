```python
#Getting Percentage of missing values in each column
#df.shape[0] represents total number of rows in df
null_var = df.isnull().sum()/df.shape[0]*100
null_var
```

```python
columns_to_drop = null_var[null_var > 17].keys()
columns_to_drop
```

```python
df_drooped_col = df.drop(columns = columns_to_drop)
```

```python
df_drooped_col.shape
```

```python
df_drooped_col.isnull().sum().sum()
```

```python
sns.heatmap(df_drooped_col.isnull())
```

```python
df_drooped_col.head(10)
```

```python
#Removes all rows that contain at least one missing value(i.e NaN)
df_drooped_NaN_rows = df_drooped_col.dropna()
df_drooped_NaN_rows.shape
```

```python
df_drooped_NaN_rows.isnull().sum().sum()
```

```python
sns.heatmap(df_drooped_NaN_rows.isnull())
```

```python
#Using distribution plot from seaborn
#It takes only numeric columns
nume_cols = list(df_drooped_NaN_rows.select_dtypes(include=['int64', 'float64']).columns)
nume_cols
```

```python
#Used to plot a histogram of observations. 
#It displays the distribution of a single variable by dividing the data into intervals called "bins"
#and counting the number of observations that fall into each bin.
plt.figure(figsize=(25,25))
for i, var in enumerate(nume_cols):
    plt.subplot(9,4,i+1)
    sns.histplot(df[var], bins=20)
    sns.histplot(df_drooped_NaN_rows[var], bins=20)
```

```python
obj_cols = list(df_drooped_NaN_rows.select_dtypes(include=['object']).columns)
obj_cols
```

```python
def cat_var_dist(var):
    return pd.concat([df[var].value_counts()/df.shape[0]*100,
            df_drooped_NaN_rows[var].value_counts()/df_drooped_NaN_rows.shape[0]*100],axis=1,
                     keys=[var+'_original', var+'_cleaned'])
```

```python
cat_var_dist('SaleType')
```
