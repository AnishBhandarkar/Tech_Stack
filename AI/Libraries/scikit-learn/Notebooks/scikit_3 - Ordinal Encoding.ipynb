# Ordinal Encoding

* It is a method of encoding categorical data especially ordinal data in to numerical data.
* Here we specify a natural order during conversion

```python
import pandas as pd
from category_encoders import OrdinalEncoder
```

```python
data = pd.read_csv('../Datasets/titanic.csv')
data.head()
```

```python
data['class'].value_counts()
```

```python
# Specify order
maplist = [{'col': 'class',
            'mapping': {'First': 0, 'Second': 1,'Third': 2}}]

ordinal_encode = OrdinalEncoder(mapping=maplist)
```

```python
data = ordinal_encode.fit_transform(data)
```
