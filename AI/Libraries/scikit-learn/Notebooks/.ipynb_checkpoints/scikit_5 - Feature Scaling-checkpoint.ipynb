# Feature Scaling

## 1.1 Standardization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('../Datasets/iris.csv')
data.head()
```

```python
plt.figure(figsize=(16,9))
for i,var in enumerate(data.columns):
    plt.subplot(3, 3,i+1)
    sns.histplot(data[var])
```

```python
# Splitting data into dependent and target variable
x = data.drop('species', axis=1)
y = data['species']
```

```python
x.head()
```

```python
y.head()
```

```python
# x has all numerical columns and all 4 columns have differnt scale. So we can apply Standardization
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
x_scaled.shape
```

```python
# Converting this numpy array back to dataframe
df_scaled = pd.DataFrame(x_scaled, columns=x.columns)
df_scaled.head()
```

```python
print("Range of Standardized Features:")
print(df_scaled.describe().loc[['min', 'max']])
```

```python
# This obtained df can be integrated with original if it has other columns
```

```python
plt.figure(figsize=(16,9))
for i,var in enumerate(df_scaled.columns):
    plt.subplot(3, 3,i+1)
    sns.histplot(df_scaled[var])
```

```python
df_scaled.sepal_length.mean()
```

```python
# It has a mean of zero and a standard deviation of one.
```

## 1.2 Normalization

```python
data = pd.read_csv('../Datasets/iris.csv')
data.head()
```

```python
# Splitting data into dependent and target variable
x = data.drop('species', axis=1)
y = data['species']
```

```python
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
x_scaled.shape
```

```python
# Converting this numpy array to df
df_scaled = pd.DataFrame(x_scaled, columns=x.columns)
df_scaled.head()
```

```python
print(df_scaled.describe().loc[['min', 'max']])
```

```python
# In Normalization the range is always between 0 and 1
```
