# 2. Importing Datasets

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris
```

```python
# Now we have to convert that to a dataframe
import pandas as pd
iris_df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_df['target'] = pd.Series(iris['target'])
```

```python
iris_df.head()
```

```python
iris_df['target'].unique()
```

```python
iris_df.shape
```

# 3. Following the above map to select best estimator

```python
# After following map, it is best to select Linear SVC (Support Vector Classification)

# Load required modules
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# create dependent and independent variables
x = iris_df.drop('target', axis=1)
y = iris_df['target']

# Split dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create Model
model = svm.LinearSVC()

# Train model
model.fit(x_train, y_train)

# Make Predictions
y_pred = model.predict(x_test)

# Evaluation

#Training accuracy
print('Training Accuracy:', model.score(x_train, y_train)* 100)
print('Testing Accuracy:', accuracy_score(y_test, y_pred) * 100)
```

```python
# As we can see from above results LinearSVC is suitable for the data we have.
# Now lets consider a situation where LinearSVC doesnt work.
# So, as per map, the next suitable algorithm is KNN. So lets try that.
```

```python
# After following map, it is best to select Linear SVC (Support Vector Classification)

# Load required modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# create dependent and independent variables
x = iris_df.drop('target', axis=1)
y = iris_df['target']

# Split dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create Model
model = KNeighborsClassifier()

# Train model
model.fit(x_train, y_train)

# Make Predictions
y_pred = model.predict(x_test)

# Evaluation

#Training accuracy
print('Training Accuracy:', model.score(x_train, y_train)* 100)
print('Testing Accuracy:', accuracy_score(y_test, y_pred) * 100)
```
