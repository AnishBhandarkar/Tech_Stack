# Scikit-learn (sklearn)

## 1. Import the necessary modules and libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

## 2. Load and preprocess the data.

```python
data = pd.read_csv('../Datasets/heart.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data.duplicated().sum()
```

```python
# Before removing duplicate values
```

```python
plt.figure(figsize=(16, 14))
for i, var in enumerate(data.columns):
    plt.subplot(5, 3, i+1)
    sns.scatterplot(data[var], label=var, color='red', alpha=0.5)
    plt.legend()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
# After removing duplicate values
```

```python
plt.figure(figsize=(16, 14))
for i, var in enumerate(data.columns):
    plt.subplot(5, 3, i+1)
    sns.scatterplot(data[var], label=var, color='red', alpha=0.5)
    plt.legend()
```

```python
data.head()
```

```python
data.isnull().sum().sum()
```

```python
# Creating feature vector and target variable
x = data.drop('target', axis=1)

y = data['target']
```

## 3. Split the data into training and testing sets.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## 4. Create a machine learning model.

```python
classifier = RandomForestClassifier()
```

## 5. Train the model using the training data.

```python
classifier.fit(x_train, y_train)
```

## 6. Make predictions on the test data.

```python
y_pred = classifier.predict(x_test)
y_pred
```

## 7. Evaluate the model's performance using appropriate metrics.

```python
classifier.score(x_train, y_train)
```

```python
classifier.score(x_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
accuracy_score(y_test, y_pred) * 100 
```

```python
cf = confusion_matrix(y_test, y_pred)
cf
```

```python
sns.heatmap(cf)
```

## 8. Save the model

```python
import pickle

pickle.dump(classifier, open('RF_Model.pkl', 'wb'))
```

```python
# Loading the model
rf = pickle.load(open('RF_Model.pkl', 'rb'))
rf.score(x_test, y_test)
```
