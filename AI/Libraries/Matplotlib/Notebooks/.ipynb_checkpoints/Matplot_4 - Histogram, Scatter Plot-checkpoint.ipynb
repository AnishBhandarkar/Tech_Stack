```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-v0_8')
```

# 1) Histogram

```python
age = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]

plt.hist(age)

plt.xlabel('Ages')
plt.ylabel('People')

plt.show()
```

#### 1.1 Providing Edge color

```python
age = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]

plt.hist(age, edgecolor='black')

plt.xlabel('Ages')
plt.ylabel('People')

plt.show()
```

#### 1.2 Providing Bin

```python
age = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]

plt.hist(age, edgecolor='black', bins=5)

plt.xlabel('Ages')
plt.ylabel('People')

plt.show()
```

#### 1.3 Customized Bin

```python
age = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]

# I want to distribute the age data between the age pf 10 to 60
bins = [10, 20, 30, 40, 50, 60]

plt.hist(age, edgecolor='black', bins=bins)

plt.xlabel('Ages')
plt.ylabel('People')

plt.show()
```

# 2) Scatter plot

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y)

plt.show()
```

### 2.1 Specifying size of points

```python
plt.scatter(x, y, s=200)

plt.show()
```

### 2.2 Specifying color

```python
plt.scatter(x, y, s=150, c='#40128B')

plt.show()
```

### 2.3 Specifying markers

```python
plt.scatter(x, y, c='#1D267D', marker='+')
plt.show()
```

### 2.4 Edge color

```python
plt.scatter(x, y, c='white', edgecolor='black')
plt.show()
```

### 2.5 Controlling color intensity

```python
plt.scatter(x, y, c='#66347F')
plt.show()
```

```python
plt.scatter(x, y, c='#66347F', alpha=0.5)
plt.show()
```

### 2.6 Representing each point with different color

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, c=colors)

plt.show()
```

### 2.7 Using cmap to change color

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, c=colors, cmap='Set1')

plt.show()
```

### 2.8 Adding colorbar legend

```python
plt.scatter(x, y, c=colors, cmap='Set1')

cbar= plt.colorbar()
cbar.set_label('Satisfaction Level')

plt.show()
```

### 2.9 Representing each data point with different size

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x, y, s=sizes, c=colors, cmap='ocean', edgecolor='black')

plt.show()
```
