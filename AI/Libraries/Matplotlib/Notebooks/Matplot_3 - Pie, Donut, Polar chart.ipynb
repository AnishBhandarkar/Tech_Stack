```python
import matplotlib.pyplot as plt
```

### 1) Pie chart

```python
slices = [60, 40]

plt.pie(slices)

plt.title('My Pie Chart')

plt.show()
```

### 2)  Pie chart with Labels

```python
slices = [60, 10, 30]
labels = ['Sixty', 'Ten', 'Thirty']

plt.pie(slices, labels=labels)

plt.title('My Pie Chart')

plt.show()
```

### 3) Adding sperators (borders)

```python
slices = [60, 10, 30]
labels = ['Sixty', 'Ten', 'Thirty']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'})

plt.title('My Pie Chart')

plt.show()
```

### 4) Passing custom colors

```python
slices = [40, 10, 20, 30]
labels = ['Forty', 'Ten', 'Twenty','Thirty']

colors = ['#99d8c9', '#fc9272', '#fec44f', '#fdbb84']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor':'black'})

plt.title('My Pie Chart')

plt.show()
```

### 9) Polar plot

```python
import matplotlib.pyplot as plt
import numpy as np

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

# Convert data to angles
angles = np.linspace(0, 2 * np.pi, len(slices), endpoint=False).tolist()
angles += angles[:1]  # Repeat the first angle to close the circle

# Create a polar plot
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)

# Plot the bars
bars = ax.bar(angles, slices + slices[:1], width=0.5, bottom=0, linewidth=1, edgecolor='black')

# Apply colors to the bars
colors = ['#99d8c9', '#fc9272', '#fec44f', '#fa9fb5', '#c994c7']
for bar, color in zip(bars, colors):
    bar.set_facecolor(color)

# Set the labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# Set a title
ax.set_title('My 3D Pie Chart')

plt.show()
```

### 10) Donut chart

```python
import matplotlib.pyplot as plt

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
colors = ['#99d8c9', '#fc9272', '#fec44f', '#fa9fb5', '#c994c7']

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the outer pie chart
wedges, texts, autotexts = ax.pie(slices, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Set aspect ratio to create a perfect circle
ax.set_aspect('equal')

# Add a circle at the center to create the donut chart
center_circle = plt.Circle((0, 0), 0.7, fc='white')
ax.add_artist(center_circle)

# Set a title
plt.title('My Donut Chart')

plt.show()
```
