# Histogram

```python
Histograms are used for plotting the distribution of a single continuous variable.
```

```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('../Datasets/sales.csv')
data.dtypes
```

## 1.1 Plotting histogram

```python
sns.histplot(x='gross income', data = data)
```

# 1.2 Horizontal plot

```python
sns.histplot(y='gross income', data = data)
```

## 1.3 Changing width of bins

```python
sns.histplot(x='gross income', data = data, binwidth=10)
```

## 1.4 Specifying number of bins

```python
sns.histplot(x='gross income', data = data, bins=50)
```

```python
sns.histplot(x='gross income', data = data, bins=[10, 20, 30, 40, 50])
```

## 1.5 Combining with KDE

```python
sns.histplot(x='gross income', data = data, kde=True)
```

## 1.6 Using Hue

```python
sns.histplot(x='gross income', data = data, hue='Gender')
```

```python
# It is overlapping. So stack it (one above other)
sns.histplot(x='gross income', data = data, hue='Gender', multiple='stack')
```

## 1.7 Removing edges

```python
sns.histplot(x='gross income', data = data, element='step')
```

## 1.8 Showing only outline

```python
sns.histplot(x='gross income', data = data,  fill=False)
```

```python
sns.histplot(x='gross income', data = data, element='step', fill=False)
```

## 1.9 polygonal line anf fill to the histogram plot. 

```python
sns.histplot(x='gross income', data = data, element='poly', color='orange')
```

```python
sns.histplot(x='gross income', data = data, element='poly', hue='Payment')
```

```python
sns.histplot(x='gross income', data = data, element='poly', hue='Payment', fill = False)
```

## 2.0 Changing stat

```python
sns.histplot(x='gross income', data = data, color='orange', stat='probability')
```
