# Swarm plot

```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('../Datasets/sales.csv')
data.dtypes
```

## 1.1 For numeric variable

```python
sns.swarmplot(y = 'Total', data=data)
```

## 1.2 Categoric variable

```python
sns.swarmplot(x='Date', data=data)
```

## 1.3 Horizontal plot

```python
sns.swarmplot(x = 'Total', data=data, color='pink')
```

## 1.4 Two variable plot

```python
sns.swarmplot(x='Payment', y = 'Total', data=data)
```

```python
# Using hues

sns.swarmplot(x='Payment', y = 'Total', hue='Gender', data=data)
```

```python
# Separating hue
sns.swarmplot(x='Payment', y = 'Total', hue='Gender', data=data, dodge=True)
```

## 1.5 Marker, Size

```python
#Marker
sns.swarmplot(x='Payment', y = 'Total', data=data, palette='flag', marker='^')
```

```python
# size
sns.swarmplot(x='Payment', y = 'Total', data=data, palette='flag', marker='*', size=4)
```

## 1.6 Swarm plot with violin plot

```python
sns.swarmplot(x='Payment', y = 'Total', data=data, palette='Pastel2')
sns.violinplot(x='Payment', y = 'Total', data=data)
```

## 1.7 Swarm plot with box plot

```python
sns.swarmplot(x='Payment', y = 'Total', data=data, palette='Pastel1')
sns.boxplot(x='Payment', y = 'Total', data=data)
```
