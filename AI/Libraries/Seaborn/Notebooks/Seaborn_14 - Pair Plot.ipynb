# Pair Plot

```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('../Datasets/iris.csv')
data.head()
```

## 1.1 Creating a pair plot

```python
sns.pairplot(data=data)
```

## 1.2 Changing diagonal plot

```python
sns.pairplot(data=data, diag_kind='kde')
```

## 1.3 Changing Non diagonal plot

```python
sns.pairplot(data=data, kind='hist')
```

## 1.4 Adding hue

```python
sns.pairplot(data=data, kind='kde', diag_kind='hist', hue='species')
```

## 1.5 Creating pair plot for specific variables

```python
sns.pairplot(data=data, x_vars=['sepal_length', 'sepal_width', 'petal_width'], y_vars='species')
```

```python
sns.pairplot(data=data, x_vars=['sepal_length', 'sepal_width'], y_vars=['petal_length', 'petal_width'], kind='kde')
```

## 1.6 Showing only lower triangle of pairplot

```python
sns.pairplot(data=data, kind='reg', corner=True)
```

## 1.7 Making changes specific to diagonal and non diagonal plots separately

```python
sns.pairplot(data=data, diag_kws=dict(kde=True, color='#16FF00'), plot_kws=dict(color='#060047', marker='D', s=5))
```

## 1.8 Creating kde plot on top of pairplot

```python
pl = sns.pairplot(data=data, plot_kws=dict(color='red'), diag_kws=dict(element='poly', color='pink'))
pl.map_upper(sns.kdeplot) # For only upper
pl.map_lower(sns.kdeplot, fill=True) #For only lower
```
