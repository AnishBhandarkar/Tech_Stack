# Cat Plot

```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('../Datasets/sales.csv')
```

## 1.1 Plotting catplot (by default it gives strip plot)

```python
sns.catplot(x = 'Payment', y = 'Total', data=data)
```

## 1.2 Plotting different plots using kind

```python
#bar plot
sns.catplot(x = 'Payment', y = 'Total', data=data, kind='bar')
```

```python
#violoin plot
sns.catplot(x = 'Payment', y = 'Total', data=data, kind='violin')
```

## 1.3 Adding hue

```python
sns.catplot(x = 'Payment', y = 'Total', hue='Gender', data=data, kind='box')
```

## 1.4 Splitting hue on a separate column

```python
sns.catplot(x = 'Payment', y = 'Total', col='Gender', data=data, kind='box')
```

## 1.5  Splitting hue on a separate row

```python
sns.catplot(x = 'Payment', y = 'Total', row='Gender', data=data, kind='box')
```

## 1.6 Combining bothrow and col

```python
sns.catplot(x = 'Payment', y = 'Total', row='Gender', col='Branch',data=data, kind='swarm', palette='Pastel2')
```
