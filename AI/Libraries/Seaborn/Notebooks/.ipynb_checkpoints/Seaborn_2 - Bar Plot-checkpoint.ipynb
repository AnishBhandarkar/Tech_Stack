```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../Datasets/titanic.csv')
data.head()
```

## 1) Bar plot

```python
data.dtypes
```

```python
sns.barplot(x='class', y='age', data=data)
plt.show()
```

```python
#Adding hue
sns.barplot(x='class', y='age', data=data, hue='sex')
plt.show()
```

```python
#Horizontal plot
sns.barplot(y='class', x='age', data=data)
plt.show()
```

```python
# Adding colors
sns.barplot(y='class', x='age', hue='sex', data=data, palette='Pastel2')
plt.show()
```

```python
# Changing order
sns.barplot(x='class', y='age', data=data, palette='Pastel2', order=['First', 'Second', 'Third'])
plt.show()
```

```python
# Changing hue order
sns.barplot(x='class', y='age', hue='sex', data=data, palette='Pastel2', hue_order=['female', 'male'])
plt.show()
```

```python
#Removing error bar
sns.barplot(x='class', y='age', data=data, palette='Pastel1', errorbar=None)
plt.show()
```

```python
# Same color for all bars
sns.barplot(x='class', y='age', data=data, color='#804674', errorbar=None)
plt.show()
```

```python
# Changing color saturation
sns.barplot(x='class', y='age', data=data, palette='Pastel1', errorbar=None, saturation=0.25)
plt.show()
```

```python
sns.barplot(x='class', y='survived', data=data, palette='Dark2', errorbar=None, estimator=len)
plt.show()
```
