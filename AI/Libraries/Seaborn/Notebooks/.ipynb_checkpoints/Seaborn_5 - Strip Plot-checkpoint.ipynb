# Strip Plot

<img src='../Embedded_images/strip_plot.png' width='400'>

```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('../Datasets/sales.csv')
data.dtypes
```

## 1.1 For numerical variable

```python
sns.stripplot(y='Total', data=data)
```

## 1.2 For categorical variable

```python
sns.stripplot(x='City', data=data)
```

## 1.3 Horizontal plot

```python
sns.stripplot(x='Total', data=data)
```

## 1.4 jitter

```python
sns.stripplot(y='Total', data=data, jitter=0.3)
```

## 1.5 linewidth

```python
sns.stripplot(x='City', data=data, jitter=0.3, linewidth=1)
```

## 1.6 Plotting using two variables

```python
data.dtypes
```

```python
sns.stripplot(x='Payment', y='Total', data=data, jitter=0.3, linewidth=0.2)
```

```python
# Using Hue
sns.stripplot(x='Payment', y='Total', hue= 'Gender' ,data=data, jitter=0.3, linewidth=0.2)
```

## 1.7 separating hue

```python
sns.stripplot(x='Payment', y='Total', hue= 'Gender' ,data=data, jitter=0.3, dodge=True)
```

## 1.8 Draw strips on top of violin plot

```python
sns.stripplot(x='Payment', y='Total', data=data, jitter=0.3, palette='Pastel1')
sns.violinplot(x='Payment', y='Total', data=data)
```

## 1.9 Draw strips on top of box plot

```python
sns.stripplot(x='Payment', y='Total', data=data, jitter=0.3, palette='Set3')
sns.boxplot(x='Payment', y='Total', data=data)
```
