```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../Datasets/titanic.csv')
data.head()
```

```python
data.dtypes
```

```python
# Plotting for single category
sns.boxplot(x = 'age', data=data)
```

```python
# Horizontal plot
# Plotting for single category
sns.boxplot(y = 'age', data=data)
```

```python
# Changing width of box
sns.boxplot(y = 'age', data=data, width=0.25)
```

```python
# Changing width of line
sns.boxplot(y = 'age', data=data, linewidth=0.5)
```

```python
# plotting with a categorical variable
sns.boxplot(x= 'class', y = 'age', data=data)
```

```python
# Using hue
sns.boxplot(x= 'class', y = 'age', hue='sex', data=data)
```

```python
# showing mean value
sns.boxplot(x= 'class', y = 'age', hue='sex', data=data, showmeans=True)
```

```python
# Changing appearance of mean marker
sns.boxplot(x= 'class', y = 'age', hue='sex', data=data, showmeans=True, meanprops={'marker':'o', 'markerfacecolor':'white', 
                                                                                   'markeredgecolor':'black', 'markersize':5})
```

```python
# Horizonal box plot

# reverse x and y label

sns.boxplot(y= 'class', x = 'age', hue='sex', data=data)
```

```python
# Changing palette
sns.boxplot(y= 'class', x = 'age', data=data, palette='Set3')
```

```python
# Creating boxplot for each numeric variable in dataframe

sns.boxplot(data=data)
```

```python
plt.yscale('log')
sns.boxplot(data=data, palette='tab20b')
```
