# Scatter Plot

```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('../Datasets/iris.csv')
data.head()
```

## 1.1 Creating a bsic scatter plot 

```python
sns.scatterplot(data=data, x='sepal_length', y='sepal_width')
```

## 1.2 Using hue

```python
sns.scatterplot(data=data, x='sepal_length', y='sepal_width', hue='species', palette='turbo')
```

## 1.3 Specifying different markers instead of color

```python
#For that use style instead of hue
sns.scatterplot(data=data, x='sepal_length', y='sepal_width', style='species',
                markers={'setosa':'^', 'versicolor':'<', 'virginica':'>'})
```

## 1.3 Specifying different markers and colors at same time

```python
#For that use both hue and style instead of hue
sns.scatterplot(data=data, x='sepal_length', y='sepal_width', style='species', hue='species',
                markers={'setosa':'^', 'versicolor':'<', 'virginica':'o'})
```

## 1.4 Using numeric variable to provide size

```python
sns.scatterplot(data=data, x='sepal_length', y='sepal_width', hue='species', size='petal_width')
```

## 1.5 Changing size of markers

```python
sns.scatterplot(data=data, x='sepal_length', y='sepal_width', hue='species', s=200)
```

## 1.6 Changing color and edgecolor

```python
sns.scatterplot(data=data, x='sepal_length', y='sepal_width', color='red', edgecolor='black', s=400)
```
