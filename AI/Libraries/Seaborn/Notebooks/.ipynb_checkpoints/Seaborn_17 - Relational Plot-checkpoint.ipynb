# Rel Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Datasets/titanic.csv')
data.head()
```

```python
data.dtypes
```

## 1.1 Basic rel plot

```python
sns.relplot(data=data, x='fare', y='age')
```

## 1.2 Changing to line plot

```python
sns.relplot(data=data, x='fare', y='age', kind='line', errorbar=None)
```

## 1.3 Changing hue

```python
sns.relplot(data=data, x='fare', y='age', hue='embark_town')
```

## 1.4 Splitting them in separate columns

```python
sns.relplot(data=data, x='fare', y='age', col='embark_town')
```

## 1.5 Splitting them in separate rows

```python
sns.relplot(data=data, x='fare', y='age', row='sex')
```

## 1.6 Specifying number of columns

```python
sns.relplot(data=data, x='fare', y='age', hue='sex', col='embark_town', col_wrap=2)
```

## 1.7 Using both rows and cols with hue

```python
#col_wrap cannot be used when row is specified
sns.relplot(data=data, x='fare', y='age', hue='sex', col='embark_town', row='class', palette='ocean')
```
