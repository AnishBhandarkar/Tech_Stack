# Displot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Datasets/sales.csv')
data.dtypes
```

## 1.1 Default Displot

```python
sns.displot(x = 'Rating', data=data)
```

## 1.2 Displot with KDE and Rug

```python
sns.displot(x = 'Rating', data=data, kde=True, rug=True)
```

## 1.3 Changing KDE and Rug attributes

```python
sns.displot(x = 'Rating', data=data, rug=True, kde=True, rug_kws={'height':0.1}, kde_kws={'bw_adjust':0.12})
```

## 1.4 Using hue

```python
sns.displot(x = 'Total', data=data, kde=True, rug=True, hue='Gender')
```

## 1.5 Stacking hue

```python
sns.displot(x = 'Total', data=data, kde=True, rug=True, hue='Gender', multiple='stack')
```

```python
sns.displot(x = 'Total', data=data, kde=True, rug=True, hue='Gender', element='poly')
```

## 1.6 Having hue in separate column

```python
sns.displot(x = 'Total', data=data, kde=True, rug=True, col='Gender')
```

## 1.7 Having plots in different rows

```python
sns.displot(x = 'Total', data=data, kde=True, rug=True, col='Gender', row='City', color='#4942E4')
```

## 1.8 Horizontal Plots

```python
sns.displot(y = 'Total', data=data, kde=True, rug=True, row='City', color='#BE5A83')
```

## 1.9 Using different plots using kind

```python
sns.displot(y = 'Total', data=data, kind='ecdf', col='Gender', row='City',  color='#19A7CE')
```
