# Heat Map

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Datasets/sales.csv')
data.dtypes
```

```python
data.shape
```

```python
data = data.loc[0:19, ['Unit price', 'Quantity', 'Tax 5%', 'Total', 'cogs', 'gross margin percentage', 'gross income', 'Rating']]
```

```python
data.shape
```

## 1.1 Creating a heatmap

```python
plt.figure(figsize=(10, 10))
sns.heatmap(data)
```

## 1.2 Annotating each box with values

```python
plt.figure(figsize=(10, 10))
sns.heatmap(data, annot=True, fmt='.0f')
```

## 1.3 Changing attributes

```python
plt.figure(figsize=(10, 10))
sns.heatmap(data, annot=True, fmt='.0f',
           annot_kws=dict(size=15, weight='bold'),
           linewidth=0.5,
           linecolor='white',
           cmap='Accent')
```

## 1.4 Plooting heatmap using correlations 

```python
data.corr()
```

```python
sns.heatmap(data.corr())
```

```python
sns.heatmap(data.corr(), vmin=-1, vmax=1, center=0, cmap='PuBuGn', annot=True, fmt='.1f', linewidth=0.5, linecolor='black',
           annot_kws=dict(size=10, weight='bold', color='black'))
```
