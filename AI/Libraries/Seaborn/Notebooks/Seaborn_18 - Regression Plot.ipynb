# Reg Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../Datasets/tips.csv')
data.head()
```

## 1.1 Creating basic reg plot

```python
sns.regplot(data=data, x='total_bill', y='tip')
```

## 1.2 Changing color and marker

```python
sns.regplot(data=data, x='total_bill', y='tip', color='#624F82', marker='+')
```

## 1.3 Changing scatter and line specifications

```python
sns.regplot(data=data, x='total_bill', y='tip', 
            scatter_kws=dict(color='#5800FF', s=100, alpha=0.5), 
            line_kws=dict(color='#FF0060', linestyle='--'))
```

## 1.4 Removing shaded area on line

```python
sns.regplot(data=data, x='total_bill', y='tip', 
            scatter_kws=dict(color='#5800FF', alpha=0.5), 
            line_kws=dict(color='#FF0060'),
            ci=None)
```
