```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../Datasets/sales.csv')
data.head()
```

```python
data.shape
```

```python
data.dtypes
```

```python
sns.violinplot(x='Total', data=data)
```

```python
# Vertical Violin plot
sns.violinplot(y='Total', data=data)
```

```python
# Using multiple variables (one should be numeric)
sns.violinplot(x = 'Payment' ,y='Total', data=data)
```

```python
# Using two categorical variables (Done using hue)

sns.violinplot(x = 'Payment' ,y='Total', hue='Gender', data=data, hue_order = ['Male', 'Female'])
```

```python
# Combining hue in one 
sns.violinplot(x = 'Payment' ,y='Total', hue='Gender', data=data, split=True)
```

```python
# Change boxplot inside violin to lines
sns.violinplot(x = 'Payment' ,y='Total', data=data, inner='quartile')
```

```python
# Draw line for each observation in violin
sns.violinplot(x = 'Payment' ,y='Total', data=data, inner='stick')
```

```python
# Smoothening the violin
sns.violinplot(x = 'Payment' ,y='Total', data=data, bw=0.23)
```

```python
# Cut out extreme values
sns.violinplot(x = 'Payment' ,y='Total', data=data, bw=0.6, cut=0)
```

```python
# Adding palette
sns.violinplot(x = 'Payment' ,y='Total', data=data, bw=0.34, palette='Pastel1')
```
