# Joint Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Datasets/iris.csv')
data.dtypes
```

```python
data.head()
```

```python
data.shape
```

## 1.1 Plotting basic jointplot

```python
sns.jointplot(x='sepal_length', y='petal_length', data=data)
```

## 1.2 Dufferent plots using kind

```python
sns.jointplot(x='sepal_length', y='petal_length', data=data, kind='hex')
```

## 1.3 Using hue

```python
sns.jointplot(x='sepal_length', y='petal_length', data=data, kind='kde', hue='species', palette='BuPu')
```

## 1.4 Styling individual plots

```python
#Changes in joint plot : joint_kws
#Changes in marginal plot : marginal_kws
sns.jointplot(x='sepal_length', y='petal_length', data=data, kind='reg', palette='BuPu',
             joint_kws=dict(marker='*', color='red'),
             marginal_kws=dict(color='pink', element='poly'))
```

## 1.5 Adding ticks to marrginal distributions

```python
sns.jointplot(x='sepal_length', y='petal_length', data=data, kind='reg', color='#34FA23', marginal_ticks=True)
```

## 1.6 Changing height

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=data, kind='kde', fill=True, color='#BC2A23', marginal_ticks=True,
              height=3)
```

## 1.7 Drawing kde and rugplot on top of joint plot

```python
pl = sns.jointplot(x='petal_length', y='petal_width', data=data, color='#BC2E12',height=4)
pl.plot_joint(sns.kdeplot, color='pink')
pl.plot_joint(sns.rugplot, height=0.1, color='green')
```
