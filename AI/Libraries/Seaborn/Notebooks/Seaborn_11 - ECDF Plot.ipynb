```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Datasets/sales.csv')
data.dtypes
```

## 1.1 ECDF Plot for Numeric variable

```python
sns.ecdfplot(x='Total', data=data)
```

## 1.2 ECDF Plot for Categoric variable

```python
sns.ecdfplot(x='Payment', data=data)
```

## 1.3 Using hue

```python
sns.ecdfplot(x='gross income', data=data, hue='Gender')
```

## 1.4 Changing stat

```python
sns.ecdfplot(x='gross income', data=data, hue='Payment', stat='count')
```
