# Kernel Density EStimation (KDE) Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Datasets/sales.csv')
data.dtypes
```

## 1.1 Plotting KDE

```python
sns.kdeplot(x='Total', fill=True, data=data)
```

## 1.2 Horizontal Plot

```python
sns.kdeplot(y='Total', data=data)
```

## 1.3 KDE Plot for all numerical variables in df

```python
sns.kdeplot(data=data)
```

## 1.4 Controlling bandwidth

```python
sns.kdeplot(x='Total', data=data, bw_adjust=0.2)
```

## 1.5 Using hue

```python
sns.kdeplot(x='Total', data=data, hue='Payment')
```

## 1.6 Stacked KDE plot

```python
sns.kdeplot(x='Total', data=data, hue='Payment', multiple='stack', palette='ocean', alpha=0.5)
```

## 1.7 Bivariate KDE

```python
#Both x and y should be numerical
sns.kdeplot(x='Total', y='Quantity', data=data)
```

## 1.8 Using hue in bivariate KDE

```python
sns.kdeplot(x='Total', y='Quantity', data=data, hue='Gender', palette='Set2')
```

## 1.9 Color Representation

```python
sns.kdeplot(x='Total', y='Quantity', data=data, hue='Gender', palette='Set2', fill=True)
```

## 2.0 Specifying contours

```python
#levels attribute is used to specify number of contours
sns.kdeplot(x='Total', y='Quantity', data=data, hue='Gender', palette='ocean', fill=True, levels=5) 
```
