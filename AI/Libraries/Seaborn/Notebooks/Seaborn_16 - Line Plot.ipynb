# Line Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Datasets/rainfall.csv')
data.head()
```

```python
data.shape
```

```python
data.dtypes
```

## 1.1 Creating basicline plot

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=data, x='YEAR', y='ANNUAL')
```

## 1.2 Removing background

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=data, x='YEAR', y='ANNUAL', errorbar=None)
```

## 1.3 Changing the estimator

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=data, x='YEAR', y='ANNUAL', estimator=None, errorbar=None)
```

## 1.4 Using hue

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=data, x='YEAR', y='ANNUAL', errorbar=None, hue='SUBDIVISION', palette='Pastel1')
```
