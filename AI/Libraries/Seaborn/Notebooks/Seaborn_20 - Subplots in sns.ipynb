```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../Datasets/iris.csv')
data.head()
```

```python
plt.style.use('seaborn-v0_8')
```

```python
#Default (1, 1)
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(data=data, x='sepal_length', y='petal_length', errorbar=None)
```

```python
# 1D array
fig, ax = plt.subplots(1, 4, figsize=(17,7))
sns.histplot(ax=ax[0] ,data=data, x='sepal_length', kde=True, color='#2B3467')
sns.histplot(ax=ax[1] ,data=data, x='sepal_width',  kde=True, color='#2B3467')
sns.histplot(ax=ax[2] ,data=data, x='petal_length', kde=True, color='#2B3467')
sns.histplot(ax=ax[3] ,data=data, x='petal_width',  kde=True, color='#2B3467')
```

```python
# 1D array
fig, ax = plt.subplots(4, 1, figsize=(17,20))
sns.histplot(ax=ax[0] ,data=data, x='sepal_length', kde=True, color='#009DAE')
sns.histplot(ax=ax[1] ,data=data, x='sepal_width',  kde=True, color='#009DAE')
sns.histplot(ax=ax[2] ,data=data, x='petal_length', kde=True, color='#009DAE')
sns.histplot(ax=ax[3] ,data=data, x='petal_width',  kde=True, color='#009DAE')
```

```python
# 2D array
fig, ax = plt.subplots(2, 2, figsize=(17,10))
sns.histplot(ax=ax[0, 0] ,data=data, x='sepal_length', kde=True, color='#FF0060')
sns.boxplot(ax=ax[0, 1] ,data=data, y='sepal_length',  color='#F7EC09')
sns.histplot(ax=ax[1, 0] ,data=data, x='petal_length', kde=True, color='#FF0060')
sns.boxplot(ax=ax[1, 1] ,data=data, y='petal_length',  color='#F7EC09')
```
