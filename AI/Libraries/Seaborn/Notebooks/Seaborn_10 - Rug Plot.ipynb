# Rug Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Datasets/sales.csv')
data.dtypes
```

## 1.1 Rug plot for numerical

```python
sns.rugplot(x='gross income', data=data)
```

## 1.2 Rug plot for categorical

```python
sns.rugplot(x='Payment', data=data)
```

## 1.3 Horizontal Rug Plot

```python
sns.rugplot(y='gross income', data=data)
```

## 1.4 Rug plot for 2 varibales

```python
sns.rugplot(x='gross income', y='Total', data=data)
```

## 1.5 hue for rug plot

```python
sns.rugplot(x='gross income', y='Total', data=data, hue='Gender')
```

## 1.6 Changing height of rugs

```python
plt.figure(figsize=(10, 5))
sns.rugplot(x='gross income', y='Total', data=data, height=0.1)
```

## 1.7 Using with KDE plot

```python
sns.rugplot(x='gross income', y='Quantity', data=data, height=0.05)
sns.kdeplot(x='gross income', y='Quantity', data=data, fill=True, color='purple')
```
